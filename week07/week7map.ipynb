```python
kyoto_pop = pd.read_csv('Kyoto.csv')
```

```python
kyoto_pop.head
```

```python
kyoto_pop.info
```

```python
kyoto_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
kyoto_ku
```

```python
kyoto_ku['KEY_CODE'] = kyoto_ku['KEY_CODE'].astype(int)
```

```python
kyoto_ku.info
```

```python
kyoto_ku_pop = kyoto_ku.merge(kyoto_pop,on='KEY_CODE')
```

```python
kyoto_ku_pop
```

```python
kyoto_ku_pop.info
```

```python
kyoto_ku_pop['人口総数'] = kyoto_ku_pop['人口総数'].astype(int)
```

```python
kyoto_ku_pop.info
```

```python
kyoto_ku_pop.plot(column='人口総数',legend=True)
```
