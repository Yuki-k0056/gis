```python
ibarakidata1 = pd.read_csv('data2/tblT001100S5339.zip', encoding='cp932')
ibarakidata2 = pd.read_csv('data2/tblT001100S5340.zip', encoding='cp932')
ibarakidata3 = pd.read_csv('data2/tblT001100S5439.zip', encoding='cp932')
ibarakidata4 = pd.read_csv('data2/tblT001100S5440.zip', encoding='cp932')
ibarakidata5 = pd.read_csv('data2/tblT000846S5339.zip', encoding='cp932')
ibarakidata6 = pd.read_csv('data2/tblT000846S5340.zip', encoding='cp932')
ibarakidata8 = pd.read_csv('data2/tblT000846S5440.zip', encoding='cp932')
```

```python
ibarakidata1=ibarakidata1[1:]
ibarakidata2=ibarakidata2[1:]
ibarakidata3=ibarakidata3[1:]
ibarakidata4=ibarakidata4[1:]
ibarakidata5=ibarakidata5[1:]
ibarakidata6=ibarakidata6[1:]
ibarakidata8=ibarakidata8[1:]
```

```python
ibarakidata1=ibarakidata1.replace('*',np.nan)
ibarakidata2=ibarakidata2.replace('*',np.nan)
ibarakidata3=ibarakidata3.replace('*',np.nan)
ibarakidata4=ibarakidata4.replace('*',np.nan)
ibarakidata5=ibarakidata5.replace('*',np.nan)
ibarakidata6=ibarakidata6.replace('*',np.nan)
ibarakidata8=ibarakidata8.replace('*',np.nan)
```

```python
ibarakidata2020=pd.concat([ibarakidata1,ibarakidata2,ibarakidata3,ibarakidata4])
```

```python
ibarakidata2015=pd.concat([ibarakidata5,ibarakidata6,ibarakidata8])
```

```python
ibarakidata2015.head()
```

```python
columns=['HTKSYORI', 'HTKSAKI', 'GASSAN']
ibarakidata2015=ibarakidata2015.drop(columns=columns)
ibarakidata2020=ibarakidata2020.drop(columns=columns)
```

```python
ibarakidata = pd.merge(ibarakidata2020,ibarakidata2015,on='KEY_CODE')
```

```python
ibarakidata.info()
```

```python
ibarakidata.columns
```

```python
ibarakidata.columns[1:]
```

```python
ibarakidata[ibarakidata.columns[1:]]=ibarakidata[ibarakidata.columns[1:]].astype(float)
```

```python
ibarakidata.info()
```

```python
ibarash['KEY_CODE']=ibarash['KEY_CODE'].astype(float)
```

```python
ibarash = ibarash.merge(ibarakidata, on='KEY_CODE')
```

```python
ibarash.sample(5)
```

```python
ibarash = ibarash.to_crs(3857)
```

```python
ibarash.plot(column='T001100001')
```

```python
ax = ibarash.plot(figsize=(12,12),
         column='T001100001',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8
         )
ax.axis('off')
ax.set_title('2020人口　総数人',fontsize=24, pad=20) 
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```

```python
ax = ibarash.plot(figsize=(12,12),
         column='T001100001',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8,
         legend_kwds={'shrink':0.5},
         vmin =0,
         vmax =15000
         )
ax.axis('off')
ax.set_title('2020人口　総数人',fontsize=24, pad=20) 
ax.set_xlim(ibarash.total_bounds[0],ibarash.total_bounds[2])
ax.set_ylim(ibarash.total_bounds[1],ibarash.total_bounds[3])
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```
